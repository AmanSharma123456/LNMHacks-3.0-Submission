```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
final_df=dataframe_2.copy()
```

```python
final_df.fillna(0)
```

```python
final_df=final_df.fillna(0)
```

```python
final_df.isnull().sum()
```

```python
pred_df=final_df.copy()
```

```python
pred_df.head()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(pred_df, test_size=0.2, random_state=42)
```

```python
train_set.head()
```

```python
Train_X=train_set.drop(["SeriousDlqin2yrs"],axis=1)
Train_X.head()
```

```python
Train_Y = train_set["SeriousDlqin2yrs"].copy()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(Train_X,Train_Y)
```

```python
some_data = Train_X.iloc[6:14]
some_labels = Train_Y.iloc[6:14]
print("Predictions:\t",lin_reg.predict(some_data))
```

```python
print("Labels\t\t",list(some_labels))
```

```python
Test_X=test_set.drop(["SeriousDlqin2yrs"],axis=1)
Test_Y=test_set["SeriousDlqin2yrs"].copy()

from sklearn.metrics import mean_squared_error
loan_predictions = lin_reg.predict(Test_X)
lin_mse = mean_squared_error(Test_Y, loan_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn import linear_model
logreg = linear_model.LogisticRegression(C=1e5,max_iter=500)
```

```python
logreg.fit(Train_X, Train_Y)
```

```python
print("Predictions:\t",logreg.predict(some_data))
```

```python
print("Labels\t\t",list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error
loan1_predictions = logreg.predict(Test_X)
lin_mse = mean_squared_error(Test_Y, loan1_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
Test_X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(Train_X,Train_Y)
```

```python
Grade_predictions = tree_reg.predict(Test_X)
tree_mse = mean_squared_error(Test_Y, Grade_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, Train_X, Train_Y, scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
rmse_scores.mean()
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(Train_X,Train_Y)
```

```python
scores = cross_val_score(forest_reg, Train_X, Train_Y, scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)

rmse_scores.mean()
```

```python
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier(alpha=0.00001,epsilon=0.001,loss="epsilon_insensitive", penalty="l2", max_iter=500)
clf.fit(Train_X, Train_Y)
```

```python
Grade_predictions = clf.predict(Test_X)
tree_mse = mean_squared_error(Test_Y, Grade_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```
